### 清理nan数据

```python
for i,data in enumerate(full_data):
    data = pd.concat([data,bu_clean,prev_check1,prev_check2],axis=1, join_axes=[data.index])
    
    data['AMT_ANNUITY'] = data['AMT_ANNUITY'].fillna(0)
    #data['AMT_GOODS_PRICE'] = data['AMT_GOODS_PRICE'].fillna(0)
    #data['OCCUPATION_TYPE'] = data['OCCUPATION_TYPE'].fillna('Unknown')
    data['CNT_FAM_MEMBERS'] = data['CNT_FAM_MEMBERS'].fillna(0).astype(int)
    #for three external sources, the NAN should be colomn mean? Row mean? Or Zero?
    data['EXT_SOURCE_1'] = data['EXT_SOURCE_1'].fillna(0)
    data['EXT_SOURCE_2'] = data['EXT_SOURCE_2'].fillna(0)
    data['EXT_SOURCE_3'] = data['EXT_SOURCE_3'].fillna(0)
    data['BU_ACTIVE_CREDIT'] = data['BU_ACTIVE_CREDIT'].fillna(0).astype(int)
    data['BU_BADDEBT_CREDIT'] = data['BU_BADDEBT_CREDIT'].fillna(0).astype(int)
    data['BU_CLOSE_CREDIT'] = data['BU_CLOSE_CREDIT'].fillna(0).astype(int)
    data['BU_SOLD_CREDIT'] = data['BU_SOLD_CREDIT'].fillna(0).astype(int)
    data['BU_MAX_DAY_OVERDUE'] = data['BU_MAX_DAY_OVERDUE'].fillna(0).astype(int)
    data['BU_SUM_DEBT'] = data['BU_SUM_DEBT'].fillna(0)
    data['BU_SUM_OVERDUE'] = data['BU_SUM_OVERDUE'].fillna(0)
    data['AMT_MAX_APPROVED'] = data['AMT_MAX_APPROVED'].fillna(0)
    data['AMT_MAX_REFUSED'] = data['AMT_MAX_REFUSED'].fillna(0)
    data['AMT_MIN_APPROVED'] = data['AMT_MIN_APPROVED'].fillna(0)
    data['AMT_MIN_REFUSED'] = data['AMT_MIN_REFUSED'].fillna(0)
    data['CNT_APPROVED'] = data['CNT_APPROVED'].fillna(0).astype(int)
    data['CNT_CANCELED'] = data['CNT_CANCELED'].fillna(0).astype(int)
    data['CNT_REFUSED'] = data['CNT_REFUSED'].fillna(0).astype(int)
    data['CNT_UNUSED'] = data['CNT_UNUSED'].fillna(0).astype(int)
    data['DAYS_LAST_APPROVED'] = data['DAYS_LAST_APPROVED'].fillna(1).astype(int)
    data['DAYS_LAST_DUE'] = data['DAYS_LAST_DUE'].fillna(0).astype(int)#is Zeros ok?
    data['DAYS_LAST_REFUSED'] = data['DAYS_LAST_REFUSED'].fillna(1).astype(int)
    data['DAYS_TERMINATION'] = data['DAYS_TERMINATION'].fillna(0).astype(int)#is zeros ok?
    data['AMT_INSTALMENT'] = data['AMT_INSTALMENT'].fillna(0)
    data['AMT_PAYMENT'] = data['AMT_PAYMENT'].fillna(0)
    data['AMT_TOTAL_RECEIVABLE'] = data['AMT_TOTAL_RECEIVABLE'].fillna(0)
    
    full_data[i] = data
```
